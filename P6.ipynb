```python
# While 반복문1
# 반복문을 통하여 누적을 구하는 예제
i = 0 # 반복 카운터, 초기화
sum = 0 # 누적할 변수, 초기화
while i < 100 : # 0 부터 100까지 실행, 만일 실행횟수는 100회
    i += 1 # i=0+1, i=1+1, i=2+1, i=3+1,i는 1씩 증가
    sum = sum + i # sum = 0 + 1, sum = 1 + 2, sum = 3 + 3, sum = 6 + 4 ...
                  #sum은 더한 값이 누적됨
print("sum = ", sum) #while 반복을 빠져 나와서 출력
```

```python
# While 반복문2
i = 0 # 반복 카운터, 초기화
sum = 0 # 누적할 변수, 초기화
while i < 100 : #0 부터 100까지 실행, 만일 실행횟수는 100회
    sum = sum + i  #sum은 더한 값이 누적됨
    i += 1  # i=0+1, i=1+1, i=2+1, i=3+1,i는 1씩 증가
print("sum = ", sum)
```

```python
# 제어문1
for i in range(10): # 0부터 9까지 실행
    if(i == 3): # i가 3이라면
        break # 탈출!
    print(i, ",", end = "")
```

```python
# 제어문2
for i in range(10): # 0부터 9까지 실행
    if(i == 3): # i가 3이라면
        continue # 문장 아래를 실행하지 않고 반복문의 처음을 이동후 계속 반복수행
    print(i,",",end="")
```

```python
# 제어문 예제 1
i = 0; sum = 0 # i와 sum의 값을 초기화
while True: # 문장으로 무한루프 실행
    i += 1; sum += i # i는 1씩 증가, sum은 i의 값을 계속 더해줌
    if(i == 100) : break # i가 100이라면 탈출!
print("i = ", i ,", sum = ", sum)
```

```python
# 제어문 예제 2
for i in range(1, 121) : # 1부터 120까지 수행되는 반복문
    print("*", end = "") # * 출력
    if(i % 30 == 0) : print(end = "\n") # i를 30으로 나눈 나머지가 0이라면 줄넘김
```

```python
# 다중반복구조1
# 구구단 예제
for i in range(1, 10): # 1부터 9까지 수행되는 반복문
    for j in range(1, 10): # 1부터 9까지 수행되는 반복문
        print(i, "*", j, "=", i * j) # i * j = i*j 를 출력
    print("***********") # 한 단이 끝나면 ****를 출력
```

```python
# 다중반복 구조2
for i in range(1, 10): # 1부터 9까지 실행되는 반복문
    for j in range(i, 10): # i부터 9까지 실행되는 반복문
        print("*", end = "") # *을 출력
    print(end = "\n") # j 반복문이 끝나면 줄넘김
```

```python
# 다중반복 구조3
i = 1 
while True:
    j = i
    while True:
        if(j == 10): break # j가 10이라면
        print("*", end = "") 
        j += 1 # j에 1 증가
    if(i == 10): break # i가 10이라면
    print(end = "\n") # 줄넘김
    i += 1 # i에 1 증가
```

```python
# 심화실습1
# 달력 인쇄, 윤달은 없는 것으로 가정
Mon = 1
for LastDay in [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31] : 
    # 매달의 마지막날을 입력한 리스트
    print(Mon, "월 :", end = "") # 몇 월인지 출력
    for Day in range(1, LastDay + 1) :
    # 날짜들을 출력하기 위해서 1부터 리스트 값까지 실행되는 반복문
        print(Day, end = ",") # 일을 출력 
    print(end = "\n") # 반복문이 끝나면 줄넘김
    Mon += 1 # mon에 1씩 증가
```

```python
# 심화실습2
# 구구단 인쇄
for i in range(2, 10, 3): # 2부터 9까지 3씩 증가하는 반복문
    for j in range(2, 10): # 2부터 9까지 실행되는 반복문
        print(i, "*", j, "=", i * j, end = "\t") # 2단 값 출력
        print(i + 1, "*", j, "=", i * j, end = "\t")  # 3단 값 출력
        if(i + 2 != 10) : print(i + 2, "*", j, "=", i * j, end = "\n") # 4단 출력
        else : print(end="\n")
    print(end="\n")
```

```python
# 심화실습3
# 구구단 출력
for i in range(2, 5): # 2부터 4까지 실행되는 반복문
    for j in range(2, 10): # 2부터 9까지 실행되는 반복문
        print(i, "*", j, "=", i * j, end = "\t") # 2단 출력
        print(i + 3, "*", j, "=", i * j, end = "\t") # 5단 출력
        if(i + 6 != 10) : print(i + 6, "*", j, "=", i * j, end = "\n") # 8단 출력
        else : print(end = "\n")
    print(end = "\n")
```
